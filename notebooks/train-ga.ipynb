```python
from matplotlib import pyplot as plt
import numpy as np
import csv
```

```python
with open("../models/ga/remote4/pop200-runs3-fitness.csv", "rt") as f:
    reader = csv.reader(f)
    data = [[float(fitness) for fitness in row] for row in reader]

x,y = zip(
    *(
        (generation, fitness) for generation, fitness_vals in enumerate(data) for fitness in fitness_vals[:20]
    )
)
plt.plot(x, y, '.', alpha=0.1)
plt.show()
```

```python
N = 20
plt.plot(np.convolve([np.mean(gen) for gen in data], np.ones(N), mode="valid")/N, label="mean")
plt.plot(np.convolve([np.median(gen) for gen in data], np.ones(N), mode="valid")/N, label="median")
plt.plot(np.convolve([np.max(gen) for gen in data], np.ones(N), mode="valid")/N, label="max")
plt.legend()
plt.show()
```
