```python
from matplotlib import pyplot as plt
import csv
```

```python
with open("../models/ga/pop200-runs3-fitness.csv", "rt") as f:
    reader = csv.reader(f)
    data = [[float(fitness) for fitness in row] for row in reader]

x,y = zip(
    *(
        (generation, fitness) for generation, fitness_vals in enumerate(data) for fitness in fitness_vals
    )
)
plt.scatter(x, y)
plt.show()

plt.violinplot(data, showmeans=True, showmedians=False)
plt.show()
```
